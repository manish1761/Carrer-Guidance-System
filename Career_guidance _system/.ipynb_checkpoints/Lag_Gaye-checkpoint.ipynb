# Model Selection

# 1. Decision Tree

```python
#Decision Tree Classifier
from sklearn import tree
from sklearn.metrics import accuracy_score

def Decision_Tree(X_train,X_test,y_train,y_test):
    DT_model = tree.DecisionTreeClassifier()
    DT_model = DT_model.fit(X_train, y_train)
    
    # Prediction
    y_pred = DT_model.predict(X_test)
    y_test_arr=y_test['Suggested Job Role']
    Final = pd.DataFrame({'Predicted':y_pred, 'Actual':y_test_arr})
    Final.reset_index()
    
    accuracy = accuracy_score(y_test,y_pred)
    accuracy = accuracy*100
    #print("accuracy=",accuracy*100)

    #Number of correct predicted
    comparison_column = np.where(Final['Predicted'] == Final['Actual'], 1, 0)
    Number_of_correct_predictions=list(comparison_column).count(1)
    #print("Number_of_correct_predictions : ",((Number_of_correct_predictions)/4000)*100,"%")
    accuracy_DT=Number_of_correct_predictions
    #print(accuracy_DT)
    return DT_model
```

# 2. Support Vector Machine

```python
# Run svm with default hyper parameters
# Default hyperparameter means C=1.0, kernel=rbf and gamma=auto among other parameters
# import SVC classifier
from sklearn.svm import SVC

# import metrics to compute accuracy
from sklearn.metrics import accuracy_score

def S_V_M(X_train,X_test,y_train,y_test):
    
    # instantiate classifier with default hyperparameters
    svc=SVC() 

    # Ensure that the target variables are 1-dimensional numpy arrays
    y_train = y_train.to_numpy().ravel()
    y_test = y_test.to_numpy().ravel()

    # fit classifier to training set
    svc.fit(X_train,y_train)

    # make predictions on test set
    y_pred=svc.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)*100
    # compute and print accuracy score
    #print('Model accuracy score with default hyperparameters: {0:0.4f}'.format(accuracy_score(y_test, y_pred)*100))
    
    return svc
```

# XGBOOST

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def XG_BOOST(X_train,X_test,y_train,y_test):
    # fit model on training data
    model = XGBClassifier()
    model.fit(X_train, y_train)

    # make predictions on the test set (assuming you have X_test and y_test)
    y_pred = model.predict(X_test)

    # compute and print accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    #print(f'Model accuracy score: {accuracy * 100:.4f}')
    
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    #print(accuracy)
    
    return model
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, Normalizer

x_new = ['90', '90', '89', '78', '89', '89', '99', '89', '6', '8', '9', '9', '9', 'yes', 'python', 'medium', 'data science', 'yes', 'comedy', 'technical', 'yes']

# Convert string elements to integers for label encoding
x_new_encoded = []
for i in range(len(x_new)):
    if i < 14 or i == 18:
        try:
            x_new_encoded.append(int(x_new[i]))
        except ValueError:
            x_new_encoded.append(x_new[i])
    else:
        x_new_encoded.append(x_new[i])

# Convert to NumPy array for label encoding
x_new_array = np.array(x_new_encoded).reshape(1, -1)

# Label encoding
label_encoder = LabelEncoder()
x_new_encoded = label_encoder.fit_transform(x_new_array.ravel())

# Reshape back to the original shape
x_new_encoded = x_new_encoded.reshape(x_new_array.shape)

# Normalizing the data
normalized_input_data = Normalizer().fit_transform(x_new_encoded)

print(normalized_input_data)
```
